```python
import json

def line(str):
    print(100*str)

def heading(str):
    print("                            "+str)

heading("INVENTARY MANAGEMENT-Add products")
line("-")
while(1):
    print("\nEnter 1 to Add products")
    print("Enter 0 to Exit")
    choice=input("\nEnter Your Choice : ")

    if(choice=='1'):
        record_file = open("record.json",'r')
        record_data = record_file.read()
        record_file.close()
        records= json.loads(record_data)
        prod_id =input("\nEnter the product id : ")
        
        if prod_id in records:
            product_qn = int(input("Enter quantity : "))
            records[prod_id]['qn']= records[prod_id]['qn'] + product_qn
            print("\n")
            line("-")
            print(str(product_qn) + " "+ records[prod_id]['name'] +' added')
            print('Total '+records[prod_id]['name'] +' = '+ str(records[prod_id]['qn']))
            line("-")
        
        else:
            product_name = input("Enter the name : ")
            product_pr = float(input("Enter the price : "))
            product_qn = int(input("Enter quantity : "))
            records[prod_id] = {'name': product_name, 'pr': product_pr, 'qn': product_qn}
            print("\n")
            line("-")
            print(str(product_qn) + " "+ records[prod_id]['name'] +' added')
            line("-")
            
        
        all_data = json.dumps(records)
        record_file = open("record.json",'w')
        record_file.write(all_data)
        record_file.close()
        
    elif(choice=='0'):
        break

    else:
        print("Invalid Choice!!")
```
