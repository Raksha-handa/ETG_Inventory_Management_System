```python
import json
import pandas as pd
import datetime

def line(str):
    print(100*str)

def heading(str):
    print("                            "+str)

now = datetime.datetime.now()
heading("INVENTARY MANAGEMENT SYSTEM")
line("-")
while(1):
    print("\nEnter 1 for the list of Products")
    print("Enter 2 to Puchase the Products")
    print("Enter 3 to See Sales")
    print("Enter 0 to Exit")
    choice=input("\nEnter Your Choice : ")
    
    if(choice=='1'):
        record_file = open("record.json",'r')
        record_data = record_file.read()
        records= json.loads(record_data)
        line("-")
        print("\tLIST OF PRODUCTS")
        line("-")
        print (str(pd.DataFrame.from_dict(records).T))
        record_file.close()

    elif(choice=='2'):
        record_file = open("record.json",'r')
        record_data = record_file.read()
        record_file.close()
        records= json.loads(record_data)
        ui_prod  = input("\nEnter the product_Id: ")
        ui_quant = int(input("Enter the quantity: "))
        
        if(records[ui_prod]['qn']>=ui_quant):
            line("-")
            print("\tBILL")
            line("-")
            print("\nProduct: ", records[ui_prod]['name'])
            print("Price: ", records[ui_prod]['pr'])
            print("Total Amount: ", records[ui_prod]['pr'] * ui_quant)
            line("-")
            print("Thank You, Please Buy Again")
            line("*")
            records[ui_prod]['qn'] = records[ui_prod]['qn'] - ui_quant

            record_file = open("record.json",'w')
            all_data = json.dumps(records)
            record_file.write(all_data)
            record_file.close()

            #Sales file
            sales_file = open("sales.json",'r')
            sales_data= sales_file.read()
            sales_file.close()
            record= json.loads(sales_data)
            record[len(record)+1] = {'Date':now.strftime("%d-%m-%Y"),'Time':now.strftime("%H:%M:%S"),'name':records[ui_prod]['name'], 'pr':records[ui_prod]['pr'], 'qn':ui_quant}
            sales_file = open("sales.json",'w')
            all_data_sales = json.dumps(record)
            sales_file.write(all_data_sales)
            sales_file.close()
        else:
            print("\n\t\t We have only "+str(records[ui_prod]['qn'])+" Product!!")

  
    elif(choice=='3'):
        sales_file = open("sales.json",'r')
        sales_data = sales_file.read()
        records= json.loads(sales_data)
        line("-")
        print("\t\tSALES")
        line('-')
        print (str(pd.DataFrame.from_dict(records).T))
        sales_file.close()

    
    elif(choice=='0'):
        break


    else:
        print("\n**Invalid Choice**\n")
```
